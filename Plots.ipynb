```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df1=pd.read_csv('/content/Rootsift Quarry(nfeatures=550,contrastThreshold=0.02,edgeThreshold=3)(0.35).csv')
df1['ParameterName'] = "A"
df1['Parameter'] = 'nfeatures=550,contrastThreshold = 0.02, edgeThreshold = 3'
df2=pd.read_csv('/content/Rootsift Quarry(nfeatures=650,contrastThreshold=0.02,edgeThreshold=3)(0.35).csv')

df2['Parameter'] = 'nfeatures=650,contrastThreshold = 0.02, edgeThreshold = 3'
df2['ParameterName'] = "B"
df3=pd.read_csv('/content/Rootsift Quarry(nfeatures=750,contrastThreshold=0.02,edgeThreshold=3)(0.35).csv')
df3['ParameterName'] = "C"
df3['Parameter'] = 'nfeatures=750,contrastThreshold = 0.02, edgeThreshold = 3'
df=pd.concat([df1, df2, df3])
df_match = pd.DataFrame(data=df)
df_match
import seaborn as sns
sns.set_theme(style='whitegrid')


# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df_match, kind="bar",
    x="ParameterName", y="F-Score", hue="Parameter" ,
    ci="sd", palette="Spectral", alpha=.9, height=8, aspect=0.9
)
g.despine(left=True)
g.set_axis_labels("Parameter Combinations ", "F-Score")
g.legend.set_title("Parameter Combinations Thresholds")
g.fig.suptitle("F-Score for ROOTSIFT for different parameter combinations")
g.savefig('F-Score(Quarry ROOTSIFT without image enhancement)(0.35).png')
import seaborn as sns
sns.set_theme(style='whitegrid')


# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df_match, kind="bar",
    x="ParameterName", y="1 - Precision Rate of Matches", hue="Parameter" ,
    ci="sd", palette="Spectral", alpha=.9, height=8, aspect=0.9
)
g.despine(left=True)
g.set_axis_labels("Parameter Combinations ", "1 - Precision Rate of Matches")
g.legend.set_title("Parameter Combinations Thresholds")
g.fig.suptitle("1 - Precision Rate of Matches for ROOTSIFT for different parameter combinations")
g.savefig('1 - Precision Rate of Matches(Quarry ROOTSIFT)(0.35).png')
import seaborn as sns
sns.set_theme(style='whitegrid')


# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df_match, kind="bar",
    x="ParameterName", y="Recall Rate of Matches", hue="Parameter" ,
    ci="sd", palette="Spectral", alpha=.9, height=8, aspect=0.9
)
g.despine(left=True)
g.set_axis_labels("Parameter Combinations ", "Recall Rate of Matches")
g.legend.set_title("Parameter Combinations Thresholds")
g.fig.suptitle("Recall Rate of Matches for ROOTSIFT")
g.savefig('Recall Rate of Matches(Quarry ROOTSIFT)(0.35).png')
import seaborn as sns
sns.set_theme(style='whitegrid')


# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df_match, kind="bar",
    x="ParameterName", y="Number of Good Matches", hue="Parameter" ,
    ci="sd", palette="Spectral", alpha=.9, height=8, aspect=0.9
)
g.despine(left=True)
g.set_axis_labels("Parameter Combinations ", "Number of Good Matches")
g.legend.set_title("Parameter Combinations Thresholds")
g.fig.suptitle("Number of Good Matches for ROOTSIFT for different parameter combinations")
g.savefig('Number of Good Matches(Quarry ROOTSIFT)(0.35).png')
import seaborn as sns
sns.set_theme(style='whitegrid')


# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df_match, kind="bar",
    x="ParameterName", y="Number of Total Matches", hue="Parameter" ,
    ci="sd", palette="Spectral", alpha=.9, height=8, aspect=0.9
)
g.despine(left=True)
g.set_axis_labels("Parameter Combinations ", "Number of Total Matches")
g.legend.set_title("Parameter Combinations Thresholds")
g.fig.suptitle("Number of Total Matches for ROOTSIFT for different parameter combinations")
g.savefig('Number of Total Matches(Quarry ROOTSIFT)(0.35).png')
```
